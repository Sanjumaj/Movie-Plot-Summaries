```
from pyspark.sql import SparkSession
from pyspark.sql.functions import log
from math import log
text = sc.textFile("/FileStore/tables/plot_summaries.txt")
# text.collect()
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')

nltk.download('stopwords')


#total_count = words_wostop.map(lambda x: (x[0],len(x)-1))



```

```
#Words without stop words
#To remove stop words
stop_words = set(stopwords.words('english'))

words = text.map(lambda x : word_tokenize(x))
words_filtered = words.map(lambda x: [word for word in x if word not in stop_words])

words_wostop = words_filtered.map(lambda sublist: list(filter(lambda word: len(word) > 2, sublist)))
words_wostop.collect()
```

```
#Calculate term frequency
kv = words_wostop.flatMap(lambda x: [((x[0], curr), 1) for curr in x[1:]])

term_frequency= kv.reduceByKey(lambda x,y : x+y)
term_frequency.collect()
```

```
#reconstructing termfrequency to join movies data
tf_movies =term_frequency.map(lambda x: (x[0][0],(x[0][1],x[1])))
tf_movies.collect()
```

```
#collecting data regarding movie ID and name
movie_metadata = sc.textFile("/FileStore/tables/movie_metadata.tsv")
split_lines = movie_metadata.map(lambda line: line.split("\t"))
split_lines.collect()
movie_info = split_lines.map(lambda x: (x[0],x[2]))
movie_info.collect()
```

```
#join the movies data and term frequency structure like ('docnum',((word,termfrequecy),'movie name'))

termfrequency_w_moviename = tf_movies.join(movie_info)
termfrequency_w_moviename.collect()
```

```
#Restructuring term frequency structure like ('word',('docnum','termfrequency'))
term_frequency_w_word_as_key = termfrequency_w_moviename.map(lambda x: (x[1][0][0],(x[0],x[1][0][1],x[1][1])))
term_frequency_w_word_as_key.collect()
```

```
#Calculating document frequency structure ('word',count)
df_map = term_frequency.map(lambda x: (x[0][1],1))
df = df_map.reduceByKey(lambda x,y : x+y)
df.collect()
```

```
#calculating inverse document frequency, N=5, structure of idf('word','idfvalue')
N=42306
idf = df.map( lambda x:(x[0],log(N/x[1])))
idf.collect()
```

```
#Join term frequency and idf 
tf_idf_values= term_frequency_w_word_as_key.join(idf)
tf_idf_values.collect()
```

```
#tf*idf values structure like ('word',('docnum','name',tf*idf))
tfidf_values = tf_idf_values.map(lambda x: (x[0],(x[1][0][0],x[1][0][2],x[1][1]*x[1][0][1])))
tfidf_values.collect()
```

```
#search terms

search_terms_rdd = sc.textFile("/FileStore/tables/search-2.txt")
search_terms_rdd.collect()
```

```
import math
for line in search_terms_rdd.collect():
    print("The movies according to the search word '{}'".format(line))
    terms_data = word_tokenize(line)
    if(len(terms_data) == 1):
        result_1 = tfidf_values.filter(lambda x: x[0]==terms_data[0]).map(lambda x: (x[1][1],x[1][2]))
        result = result_1.sortBy(lambda x:-x[1]).take(10)
    else:
        terms= sc.parallelize(terms_data)
        line_map =terms.map(lambda x :(x,1))
    #Term frequency for search query structure like ('word',count)
        line_tf= line_map.reduceByKey(lambda x,y:x+y)
    #Inverse document frequency for search query structure like('word',(count,idf))
        line_idf_tf= line_tf.join(idf)
    #Inverse document frequency for search query structure like('word',tf*idf,idf)
        line_idf= line_idf_tf.map(lambda x: (x[0],(x[1][0]*x[1][1],x[1][1])))
    #join above table and tfidftable structure like ('word',('doc','name','tf_idf_doc'),(tf_idf_que,idf))
        joined=tfidf_values.join(line_idf)
    #structure is ('movie','word','multipliedtfidf','idf')
        multipliedvalue= joined.map(lambda x: (x[1][0][1],x[0],x[1][0][2]*x[1][1][0],x[1][1][1]))
    #adding the product of all words that belong to the document structure('movie','tfidfsum')
        multiplied_vector=multipliedvalue.map(lambda x: (x[0],x[2])).reduceByKey(lambda x,y:x+y)
    #taking word and tfidf from line idf, then mapping to only tfidf
    #has a value sum of squares of all terms in the search line
        search=line_idf.map(lambda x:(x[0],x[1][0])).map(lambda x: (x[1]*x[1])).reduce(lambda x,y:x+y)
    #Taking values from join,then removing word(sqrt(doc))
        doc_value = joined.map(lambda x: (x[1][0][1],x[0],x[1][0][2])).map(lambda x: (x[0],x[2]*x[2])).reduceByKey(lambda x,y: x+y).map(lambda x:(x[0],math.sqrt(x[1])))
    #joining doc_value with multiplied_vector('moviename','sqrt(sum)','vectorproduct')
        nearer= doc_value.join(multiplied_vector)
    #('moviename','sqrt(sum)','vectorproduct','searchvalue that is sum of values of query')
        nearerto=nearer.map(lambda x: (x[0],x[1][1],x[1][0],search)).map(lambda x: (x[0],x[1],x[2],math.sqrt(x[3]))).map(lambda x: (x[0],x[1]/(x[2]+x[3])))
        result= nearerto.sortBy(lambda x: -x[1]).take(10)
        
    for movie in result:
        print("\t",movie[0])
    print()
        
                                                                                                   
```
